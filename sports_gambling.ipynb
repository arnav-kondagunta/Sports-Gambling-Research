```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=data.dropna(axis=0, how='any')
data=data.drop('participant', axis=1)
data=data.drop('sghs-18_score', axis=1)
```

```python
data.describe()
```

```python
# Compute the correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, square=True, linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python
correlation_with_A = corr['pgsi_score']
print(correlation_with_A.abs().sort_values(ascending=False))
```

```python
mask = np.triu(np.ones_like(correlation_with_A, dtype=bool))
sns.heatmap(correlation_with_A.to_frame().T, annot=False, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of PGSI with Demographic & Gambling Activity Variables')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```
